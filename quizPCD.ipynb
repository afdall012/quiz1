```python
def find_max_min_pixel(image_path):
    # Baca gambar menggunakan OpenCV
    img = cv2.imread(image_path)

    # Ubah gambar menjadi skala abu-abu jika perlu
    if len(img.shape) == 3:
        img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    else:
        img_gray = img

    # Cari nilai maksimum dan minimum dari piksel
    max_pixel = img_gray.max()
    min_pixel = img_gray.min()

    return max_pixel, min_pixel

# Ganti 'nama_gambar.jpg' dengan nama gambar yang ingin Anda periksa
image_path = '/content/imagee.jpg'
max_pixel, min_pixel = find_max_min_pixel(image_path)

print("Nilai maksimum pixel:", max_pixel)
print("Nilai minimum pixel:", min_pixel)
```

```python
def normalize_image(img):
    # Normalisasi piksel
    normalized_img = img / 255.0
    return normalized_img

# Baca gambar dalam skala abu-abu (grayscale)
img = cv2.imread('/content/imagee.jpg', 0)

# Lakukan normalisasi
normalized_img = normalize_image(img)

# Konversi gambar hasil normalisasi menjadi tipe data uint8 dan dalam rentang 0-255
normalized_img_uint8 = (normalized_img * 255).astype('uint8')
```

```python
def flip_horizontal(img):
    # Lakukan flip horizontal
    flipped_img = cv2.flip(img, 1)
    return flipped_img

def flip_vertical(img):
    # Lakukan flip vertikal
    flipped_img = cv2.flip(img, 0)
    return flipped_img

# Baca gambar
img = cv2.imread('/content/imagee.jpg')

# Lakukan flip horizontal
flipped_horizontal_img = flip_horizontal(img)

# Lakukan flip vertikal
flipped_vertical_img = flip_vertical(img)
```

```python
# Fungsi untuk menambahkan noise Gaussian pada gambar
def add_gaussian_noise(img, mean=0, sigma=25):
    noise = np.random.normal(mean, sigma, img.shape).astype('uint8')
    noisy_img = cv2.add(img, noise)
    return noisy_img

# Fungsi untuk menambahkan noise salt and pepper pada gambar
def add_salt_and_pepper_noise(img, amount=0.04):
    noisy_img = img.copy()

    # Salt noise
    num_salt = np.ceil(amount * img.size * 0.5)
    coords = [np.random.randint(0, i - 1, int(num_salt)) for i in img.shape]
    noisy_img[coords[0], coords[1], :] = 255

    # Pepper noise
    num_pepper = np.ceil(amount * img.size * 0.5)
    coords = [np.random.randint(0, i - 1, int(num_pepper)) for i in img.shape]
    noisy_img[coords[0], coords[1], :] = 0

    return noisy_img

# Baca gambar
img = cv2.imread('/content/imagee.jpg')

# Tambahkan noise Gaussian
noisy_gaussian_img = add_gaussian_noise(img)

# Tambahkan noise salt and pepper
noisy_salt_pepper_img = add_salt_and_pepper_noise(img)
```

```python
# Fungsi untuk melakukan random crop
def random_crop(img, crop_size=(250, 250)):
    h, w = img.shape[:2]
    top = np.random.randint(0, h - crop_size[0] + 1)
    left = np.random.randint(0, w - crop_size[1] + 1)
    bottom = top + crop_size[0]
    right = left + crop_size[1]
    cropped_img = img[top:bottom, left:right]
    return cropped_img

# Fungsi untuk melakukan center crop
def center_crop(img, crop_size=(250, 250)):
    h, w = img.shape[:2]
    top = (h - crop_size[0]) // 2
    left = (w - crop_size[1]) // 2
    bottom = top + crop_size[0]
    right = left + crop_size[1]
    cropped_img = img[top:bottom, left:right]
    return cropped_img

# Baca gambar
img = cv2.imread('/content/imagee.jpg')

# Lakukan random crop
random_cropped_img = random_crop(img)

# Lakukan center crop
center_cropped_img = center_crop(img)
```
